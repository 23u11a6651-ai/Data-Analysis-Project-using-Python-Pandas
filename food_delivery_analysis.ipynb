```python
import pandas as pd

# Sample data
orders = pd.DataFrame({
    "order_id": [1, 2, 3, 4],
    "user_id": [101, 102, 101, 103],
    "city": ["Bangalore", "Hyderabad", "Bangalore", "Chennai"],
    "amount": [450, 300, 700, 500]
})

# Basic analysis
total_revenue = orders["amount"].sum()
avg_order_value = orders["amount"].mean()

total_revenue, avg_order_value
```
